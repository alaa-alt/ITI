#### 1) Do universities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university is one of the most preferred by students?

```python
sns.scatterplot(x=df['Percent admitted - total'], y=df['Applicants total'])
plt.show()
```

##### Answer: We can see from the graph above that the lower the acceptance rate is the higher the applicants, as this may be because students perceive it as a more prestigious university

#### 2) Do students prefer universities that have a high rate of admission?, in other words, do students prefer a university where it is easier for them to be admitted?.

##### As seen above, students prefer universities seen as "ivy league", which means they prefer universities with high admission rate. So the answer is No.

#### 3) Do students prefer public or private universities?

```python
sns.barplot(data=df, x="Control of institution", y="Applicants total", palette="tab10")
```

##### Students prefer public universities over private

#### 4) Do students prefer universities with low tuition and fees?

```python
sns.scatterplot(data = df, x="Tuition and fees, 2013-14", y="Applicants total", palette="tab10")
plt.show()
```

#### 6) Do students prefer universities from highly populated states?

```python
high_populated_states = ['California', 'Florida', 'Georgia', 'Illinois', 'Michigan','New York', 'North Carolina', 'Ohio', 'Pennsylvania', 'Texas']
```

```python
df["high_populated"] = df["State abbreviation"].apply(lambda x: "high populated" if x in high_populated_states else "low populated")
df.head()
```

```python
sns.barplot(data=df , x='high_populated' , y='Applicants total', palette='tab10')
```

##### Students prefer universities from highly populated states

#### 7) Do students prefer a university because it belongs to a state with a high GDP per capita?

```python
df["State abbreviation"].unique()
```

```python
high_gdp = ['Massachusetts', 'Connecticut', 'New York', 'California', 'New Jersey']
df["high_gdp"] = df["State abbreviation"].apply(lambda x: "high gdp" if x in high_gdp else "low gdp")
df.head()
```

```python
sns.barplot(data=df , x='high_gdp' , y='Applicants total', palette='tab10')
plt.show()
```

##### Students prefer universities from states with high GDP per capita

#### 8) Do students prefer a university based on the possibility of a higher, additional academic degree in the same university?

```python
plt.figure(figsize=(24,12))
sns.barplot(data=df, x='Highest degree offered', y='Applicants total', palette='tab10')
plt.show()
```

##### Students prefer universities that offer a higher, additional academic degree
