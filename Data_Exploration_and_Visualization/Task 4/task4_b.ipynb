```python
from selenium import webdriver
import pandas as pd
```

```python
fawzy = webdriver.Chrome()
```

```python
fawzy.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
movie_title = fawzy.find_elements("xpath", "//*/div/table/tbody/tr/td[2]/a")
titles = [item.text.strip() for item in movie_title if item.text.strip() != ""]
titles
```

```python
//*[@id="table"]/div/table[2]/tbody/tr/td[3]
```

```python
lifetime_gross = fawzy.find_elements("xpath", "//*/div/table[2]/tbody/tr/td[3]")
gross = [item.text.strip() for item in lifetime_gross if item.text.strip() != ""]
gross
```

```python
release_year = fawzy.find_elements("xpath", "//*/div/table[2]/tbody/tr/td[4]/a")
year = [item.text.strip() for item in release_year if item.text.strip() != ""]
year
```

```python
movie_data = []

for title, gross, year in zip(titles, gross, year):
    movie_data.append({"Movie Title": title, "Lifetime Gross ": gross, "Year": year})
df = pd.DataFrame(movie_data)
print(df)
```

```python
df.to_csv('movies.csv', index=False)
```
