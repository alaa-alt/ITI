```python
def GD(x,y,lr,max_iter):
    theta_0 = 0
    theta_1 = 0
    loss = []
    theta0_all = []
    theta1_all = []
    h_all = []
    for i in range(max_iter):
        print("************************iteration:", i, "*************************")
        theta0_all.append(theta_0)
        theta1_all.append(theta_1)
        h = theta_0 + theta_1 * X
        print("h(x):",h)
        h_all.append(h)
        e = h - y
        print("error vector:", e)
        n = len(y)
        MSE = (np.linalg.norm(e)**2)/n
        J = MSE/2
        print("j=", J)
        loss.append(J)
        d_theta_0 = np.sum(e)/n
        d_theta_1 = np.sum(e*X)/n
        g_vector = np.array([d_theta_0, d_theta_1])
        print("gradient vector:", g_vector)
        g_vector_norm = np.linalg.norm(g_vector)
        print("gradient vector norm", g_vector_norm)
        theta_0_new = theta_0 - lr * d_theta_0
        theta_1_new = theta_1 - lr * d_theta_1
        if g_vector_norm < 0.001:
            break
        if i>2 and abs(loss[i] - loss[i-1])<0.001:
            break
        theta_0, theta_1 = theta_0_new, theta_1_new
    return theta_0, theta_1, loss, h_all,theta0_all, theta1_all, h
```

```python
X = np.linspace(0,20)
a=-2
b=1
y=a*X+b
y
```

```python
def SGD(X,y,lr,max_iter):
    theta_0 = 0
    theta_1 = 0
    loss = []
    theta0_all = []
    theta1_all = []
    h_all = []
    n = len(y)
    index = np.random.permutation(n)
    X_shuffled = X[index]
    y_shuffled = y[index]
    for i in range(max_iter):
        print("************************epoch:", i, "*************************")
      
        avg_grad_norm = 0
        for j in range(n):
            x_i = X_shuffled[j]
            y_i = y_shuffled[j]
            h_i = theta_0 + theta_1 * x_i
            h_all.append(h_i)
            e = h_i - y_i
            
            J = ((e**2)/2)
            loss.append(J)
            d_theta_0 = e
            d_theta_1 = e*x_i

            theta_0 = theta_0 - lr * d_theta_0
            theta_1 = theta_1 - lr * d_theta_1
            theta0_all.append(theta_0)
            theta1_all.append(theta_1)

            g_vector = np.array([d_theta_0, d_theta_1])
            g_vector_norm = np.linalg.norm(g_vector)

        if g_vector_norm < 0.001:
            break
            
        if len(loss)>1 and abs(loss[-1] - loss[-2])<0.001:
            break
        print("cost=", J)
        print("gradient vector:", g_vector)
        print("gradient vector norm", g_vector_norm)
        print("error vector:", e)
        print("theta_0_new",theta_0)
        print("theta_1_new",theta_1)
        
    return theta_0, theta_1, loss, h_all,theta0_all, theta1_all
```

```python
X = np.linspace(0,20)
a=-2
b=1
y=a*X+b
y
```

```python
def MiniBatch(X,y,lr,max_iter,batch_size):
    theta_0 = 0
    theta_1 = 0
    loss = []
    theta0_all = []
    theta1_all = []
    h_all = []
    n = len(y)
    for i in range(max_iter):
        print("************************epoch:", i, "*************************")
        index = np.random.permutation(n)
        X_shuffled = X[index]
        y_shuffled = y[index]
        epoch_loss = 0
        epoch_h = []
        for j in range(0,n,batch_size):
            x_i = X_shuffled[j:j+batch_size]
            y_i = y_shuffled[j:j+batch_size]
            h_i = theta_0 + theta_1 * x_i
            epoch_h.append(h_i)
            e = h_i - y_i
            J = np.mean((e**2)/2)
            epoch_loss = epoch_loss + J

            d_theta_0 = np.mean(e)
            d_theta_1 = np.mean(e*x_i)

            theta_0 = theta_0 - lr * d_theta_0
            theta_1 = theta_1 - lr * d_theta_1

            g_vector = np.array([d_theta_0, d_theta_1])
            g_vector_norm = np.linalg.norm(g_vector)
        theta0_all.append(theta_0)
        theta1_all.append(theta_1)
        loss.append(epoch_loss / len(range(0,n,batch_size)))
        h_all.append(epoch_h)
        if g_vector_norm < 0.001:
                return theta_0, theta_1, loss, h_all,theta0_all, theta1_all
            
        if len(loss)>2 and abs(loss[-1] - loss[-2])<0.001:
            break
        print("cost=", J)
        print("gradient vector:", g_vector)
        print("gradient vector norm", g_vector_norm)
        print("error vector:", e)
        print("theta_0_new",theta_0)
        print("theta_1_new",theta_1)
        
    return theta_0, theta_1, loss, h_all,theta0_all, theta1_all
```

```python
theta_0, theta_1, loss, h_all,theta0_all, theta1_all = MiniBatch(X,y,0.0005,300,5)
```
