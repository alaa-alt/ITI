```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X = np.linspace(0,20)
a=-1
b=2
y=a*X+b
y
```

```python
plt.scatter(X, y, marker='o')
plt.grid(True)
plt.show()
```

```python
def GD(x,y,lr,max_iter):
    theta_0 = 0
    theta_1 = 0
    loss = []
    theta0_all = []
    theta1_all = []
    h_all = []
    for i in range(max_iter):
        print("************************iteration:", i, "*************************")
        theta0_all.append(theta_0)
        theta1_all.append(theta_1)
        h = theta_0 + theta_1 * X
        print("h(x):",h)
        h_all.append(h)
        e = h - y
        print("error vector:", e)
        n = len(y)
        MSE = (np.linalg.norm(e)**2)/n
        J = MSE/2
        print("j=", J)
        loss.append(J)
        d_theta_0 = np.sum(e)/n
        d_theta_1 = np.sum(e*X)/n
        g_vector = np.array([d_theta_0, d_theta_1])
        print("gradient vector:", g_vector)
        g_vector_norm = np.linalg.norm(g_vector)
        print("gradient vector norm", g_vector_norm)
        theta_0_new = theta_0 - lr * d_theta_0
        theta_1_new = theta_1 - lr * d_theta_1
        if g_vector_norm < 0.001:
            break
        if i>2 and abs(loss[i] - loss[i-1])<0.001:
            break
        theta_0, theta_1 = theta_0_new, theta_1_new
    return theta_0, theta_1, loss, h_all,theta0_all, theta1_all, h
```

```python
theta_0, theta_1, loss, h_all,theta0_all, theta1_all, h = GD(X,y,0.0005,300)
```

```python
from sklearn.metrics import r2_score
r2_score(y, h)
```

```python
def plot_loss(epochs,loss):
    plt.plot(epochs, loss, marker="o", color="tab:pink")
    plt.grid(True)
```

```python
epochs = range(len(loss))
plot_loss(epochs,loss)
plt.xlabel("epochs")
plt.ylabel("Loss")
plt.title("Loss Vs. epochs")
plt.show()
```

```python
plot_loss(theta0_all, loss)
plt.xlabel("Theta 0")
plt.ylabel("Loss")
plt.title("Learning Rate = 0.0005")
plt.show()
```

```python
plot_loss(theta1_all, loss)
plt.xlabel("Theta 1")
plt.ylabel("Loss")
plt.title("Learning Rate = 0.0005")
plt.show()
```

```python
def GD_Momentum(x,y,lr,gamma,max_iter):
    theta_0 = 0
    theta_1 = 0
    v_0 = 0
    v_1 = 0
    loss = []
    theta0_all = []
    theta1_all = []
    h_all = []
    for i in range(max_iter):
        print("************************iteration:", i, "*************************")
        theta0_all.append(theta_0)
        theta1_all.append(theta_1)
        h = theta_0 + theta_1 * X
        print("h(x):",h)
        h_all.append(h)
        e = h - y
        print("error vector:", e)
        n = len(y)
        MSE = (np.linalg.norm(e)**2)/n
        J = MSE/2
        print("j=", J)
        loss.append(J)
        d_theta_0 = np.sum(e)/n
        d_theta_1 = np.sum(e*X)/n
        v_0 = gamma*v_0 - lr*d_theta_0
        v_1 = gamma*v_1 - lr*d_theta_1
        g_vector = np.array([d_theta_0, d_theta_1])
        print("gradient vector:", g_vector)
        g_vector_norm = np.linalg.norm(g_vector)
        print("gradient vector norm", g_vector_norm)
        theta_0 = theta_0 + v_0
        theta_1 = theta_1 + v_1
        if g_vector_norm < 0.001:
            break
        if i>2 and abs(loss[i] - loss[i-1])<0.001:
            break
    return theta_0, theta_1, loss, h_all,theta0_all, theta1_all, h
```

```python
import numpy as np

def GD_NAG(X, y, lr, gamma, max_iter):
    theta_0 = 0
    theta_1 = 0
    v_0 = 0
    v_1 = 0
    loss = []
    theta0_all = []
    theta1_all = []
    h_all = []
    n = len(y)
    for i in range(max_iter):
        print("************************iteration:", i, "*************************")
        
        theta0_temp = theta_0 + gamma * v_0
        theta1_temp = theta_1 + gamma * v_1
        
        h_temp = theta0_temp + theta1_temp * X  
        e_temp = h_temp - y  
        
        print("error vector:", e_temp)
        
        MSE = (np.linalg.norm(e_temp)**2) / n
        J = MSE / 2
        print("j=", J)
        
        loss.append(J)

        d_theta_0 = np.sum(e_temp) / n
        d_theta_1 = np.sum(e_temp * X) / n

        v_0 = gamma * v_0 - lr * d_theta_0
        v_1 = gamma * v_1 - lr * d_theta_1

        theta_0 = theta_0+ v_0
        theta_1 =theta_1 + v_1
        
        h = theta_0 + theta_1 * X
        h_all.append(h)

        print("h(x):", h)
        g_vector = np.array([d_theta_0, d_theta_1])
        print("gradient vector:", g_vector)
        g_vector_norm = np.linalg.norm(g_vector)
        print("gradient vector norm", g_vector_norm)

        theta0_all.append(theta_0)
        theta1_all.append(theta_1)

        if g_vector_norm < 0.001:
            break
        if len(loss) > 1 and abs(loss[-1] - loss[-2]) < 0.001: ###the len(loss)>1 is because the first iteration won't have loss
            break

    return theta_0, theta_1, loss, h_all, theta0_all, theta1_all, h
```
